```python
import numpy as np
import matplotlib.pyplot as plt
from uncertainties import ufloat, unumpy
```

## Polarization

### P($\theta$) vs. $\theta$ Graph

```python
angles = np.arange(0,360, 20)
power = np.array([141, 33.8,1.11,58.8,179,306,379,365,271,139,33.6,0.97,57.1,178,306,378,364,268])
angles_err = np.full(shape = 18, fill_value = 0.1)
power_err = power * 0.01

plt.title(r"P($\theta$) vs. $\theta$")
plt.xlabel(r"Angle ($\theta$)")
plt.ylabel(r"Power ($\mu$W)")
plt.errorbar(angles, power, power_err, angles_err, 
            color = 'blue', marker = '.', markersize = 3, 
            markerfacecolor = 'None', linestyle = 'None', 
            ecolor = "blue", elinewidth=1, capsize=2)
plt.show()
```

### Degree of Polarization

```python
I_max = unumpy.uarray(np.max(power), np.max(power_err)) #Get maximum intensity with uncertainty
I_min = unumpy.uarray(np.min(power), np.min(power_err)) #Get minimum intensity with uncertainty

rho = (I_max-I_min) / (I_max + I_min)

print("The Degree of Polarization is: ", rho)
```
